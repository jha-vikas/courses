Please complete the code block by implementing the PageRank equation: $r_j = \sum_{i \rightarrow j} \beta \frac{r_i}{d_i} + (1 - \beta) \frac{1}{N}$

```python
nx.pagerank(Gd, max_iter=1, tol=10000, alpha=0.8)
```

```python
def one_iter_pagerank(G, beta, r0, node_id):
  # TODO: Implement this function that takes a nx.Graph, beta, r0 and node id.
  # The return value r1 is one interation PageRank value for the input node.
  # Please round r1 to 2 decimal places.
  
  r1 = 0

  N = G.number_of_nodes()
  G = G.to_directed()
    
  neigh = G.neighbors(node_id)
  for i in neigh:
    r1  = r1 + r0 / (G.out_degree(i))

  r1 = beta*r1 + (1-beta)/ N

  ############# Your code here ############
  ## Note: 
  ## 1: You should not use nx.pagerank

  #########################################

  return r1

beta = 0.8
r0 = 1 / G.number_of_nodes()
node = 1
r1 = one_iter_pagerank(G, beta, r0, node)
print("The PageRank value for node 0 after one iteration is {}".format(r1))
```
